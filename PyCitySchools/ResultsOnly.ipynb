## PyCity Schools Report

The Disrict Summary Table showed that as a whole, students in the district had a higher pass rate in reading than math by over 10 percentage points (85.81% vs 74.98%).

The School Summary Table and the filtered versions showing top and bottom performing showed even more insight on how schools performed compared to one another.
* The Top 5 performing schools are all Charter schools
* The Bottom 5 performing schools are all District schools
* Aside from the 5 Charter schools performing the best, they also are doing it with lower per student budgets (on average) than the District schools in the bottom 5.

The tables for Reading and Math Scores by Grades showed that scores stayed relatively consistent at each school from 9th to 12th grade.

The Scores by School Spending confirmed an earlier observation that the schools spending less per student are actually outperforming the schools spending more per student. In fact, the average Overall Pass % steadily decreased from over 90% for the lowest range (<$585 per student) to less than 54% for the highest range ($645 to $680 per student).

The Scores by School Size table shows that schools designated to be Small (<1000 students) or Medium (1000-2000 students) both have similar average pass rates around 90% while large schools (2000-5000 students) dropped steeply to an average pass rate around 58%.

The Scores by School Type table reinforced a previous observation that Charter schools outperform District schools with average pass rates of approximately 90% and 54% respectively.
