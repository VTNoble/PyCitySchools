```python
# Dependencies
import pandas as pd
```

```python
# Files to load
school_data_csv = 'Resources/schools_complete.csv'
student_data_csv = 'Resources/students_complete.csv'

# Read in data from csv files and store into DataFrames
school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)
```

```python
# Add a column to the school_data DataFrame with the budget_per_student
school_data['budget_per_student'] = school_data['budget'] / school_data['size']

#school_data.head()
```

```python
# Combine the data from the student and school DataFrames into a single DataFrame.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

```python
# District Summary Table
# Find values to fill table
total_schools = school_data_complete['school_name'].nunique()
total_students = school_data_complete['Student ID'].nunique()
total_budget = school_data['budget'].sum()
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()
```

```python
# Additional DataFrames needed to find values for % passing
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
passing_both = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (student_data['reading_score'] >= 70), :]

# Find values to fill rest of District Summary Tables                          
percent_math = passing_math['Student ID'].nunique() / total_students * 100
percent_reading = passing_reading['Student ID'].nunique() / total_students * 100
percent_both = passing_both['Student ID'].nunique() / total_students * 100
```

```python
# Scores by School Spending
# Create spending bins
spending_bins = [0, 585, 630, 645, 680]

# Create names for the spendings bins
spending_ranges = ['<$585', '$585-630', '$630-645', '$645-680']

# Add column and assign spending ranges to school_summary table
# Must use unformatted version of table so the budget column we are cutting is still a float datatype
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary["Per Student Budget"], spending_bins, labels=spending_ranges, include_lowest=True)

#school_summary.head()
```

<b> Scores by School Spending
* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
* Average Math Score
* Average Reading Score
* % Passing Math
* % Passing Reading
* Overall Passing Rate (Average of the above two)

```python
# Use GroupBy to filter by bins
school_spending_table = school_summary.groupby('Spending Ranges (Per Student)')

# List of columns to evaluate and include in table
by_spending_columns = ['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']

# Create table with average values per spending range
scores_by_spending = school_spending_table[by_spending_columns].mean().round(2)

# Show percent symbol on Percent Passing columns
scores_by_spending['% Passing Math'] = scores_by_spending['% Passing Math'].map("{:}%".format)
scores_by_spending['% Passing Reading'] = scores_by_spending['% Passing Reading'].map("{:}%".format)
scores_by_spending['% Overall Passing'] = scores_by_spending['% Overall Passing'].map("{:}%".format)

scores_by_spending
```

```python
# Scores by School Size
# Create size bins
size_bins = [0, 1000, 2000, 5000]

# Create names for the spendings bins
size_ranges = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']

# Add column and assign spending ranges to school_size table
# Must use unformatted version of table so the total students column we are cutting is still an int datatype
school_summary['School Size'] = pd.cut(school_summary["Total Students"], size_bins, labels=size_ranges, include_lowest=True)

#school_summary.head()
```

<b>Scores by School Size

```python
# Use GroupBy to filter by school size bins
school_size_table = school_summary.groupby('School Size')

# List of columns to evaluate and include in table
by_size_columns = ['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']

# Create table with average values per size range
scores_by_size = school_size_table[by_size_columns].mean().round(2)

# Show percent symbol on Percent Passing columns
scores_by_size['% Passing Math'] = scores_by_size['% Passing Math'].map("{:}%".format)
scores_by_size['% Passing Reading'] = scores_by_size['% Passing Reading'].map("{:}%".format)
scores_by_size['% Overall Passing'] = scores_by_size['% Overall Passing'].map("{:}%".format)

scores_by_size
```
